```python
import pandas as pd
import pandas_datareader.data as web
import datetime as dt
import numpy as np
```

## Absolute Momentum
If absolute momentum of risky asset is negative, sell risky asset and buy safe asset. <br>
If absolute momentum of risky asset is positive, buy risky asset and sell safe asset.

```python
risky = 'SPY'
safe = 'IEF'

start = dt.datetime(1970, 1, 1)
end = dt.datetime.today()

df = pd.DataFrame()

df[risky] = web.DataReader(risky, 'yahoo', start, end)['Adj Close']
df[safe] = web.DataReader(safe, 'yahoo', start, end)['Adj Close']
df.dropna(inplace=True)
df.head(40)
```

```python
df = df.resample('Y').agg(lambda x: x[-1])
df = df.iloc[:-1]
df
```

```python
for i in range(1, len(df.index)):
    df.loc[df.index[i], 'SPY SCORE'] = df.iloc[i, 0] - df.iloc[i-1, 0]
    df.loc[df.index[i], 'IEF SCORE'] = df.iloc[i, 1] - df.iloc[i-1, 1]

df
```

```python
for i in range(1, len(df.index)):
    if df.loc[df.index[i], 'SPY SCORE'] > 0:
        df.loc[df.index[i], 'SIGNAL'] = 'buy'
    else:
        df.loc[df.index[i], 'SIGNAL'] = 'sell risky and buy safe'
```

```python
df
```

```python
returns = []
posRisky = 0
posSafe = 0

for i in range(1, len(df.index)):
    date = df.index[i]
    riskyPrice = df.loc[date, risky]
    safePrice = df.loc[date, safe]
    mom_signal = df.loc[date, 'SIGNAL']

    if (mom_signal == 'buy') and (posRisky == 0):

        if posSafe > 0:  # we have safe position that we need to sell
            safeSell = safePrice
            ret = (safeSell - safeBuy)/safeBuy
            print("sell safe price: " + str(safeSell))
            print("safe position return: " + str(ret))
            print("="*15)
            returns.append(ret)
            posSafe = 0

        # buy risky asset
        riskyBuy = riskyPrice
        print("buy risky price: " + str(riskyBuy))
        posRisky = 1

    elif mom_signal == 'sell risky and buy safe' and (posRisky > 0):

        # sell risky asset
        riskySell = riskyPrice
        ret = (riskySell-riskyBuy)/riskyBuy
        print("sell risky price:" + str(riskySell))
        print("risky position return: " + str(ret))
        print("="*15)
        returns.append(ret)
        posRisky = 0

        # buy safe asset
        safeBuy = safePrice
        print("buy safe price: " + str(safeBuy))
        posSafe = 1

    if (i == len(df.index) - 1) and ((posRisky > 0) or (posSafe > 0)):
        if posRisky > 0:
            riskySell = riskyPrice
            ret = (riskySell-riskyBuy)/riskyBuy
            print("sell risky price:" + str(riskySell))
            print("risky position return: " + str(ret))
            print("="*15)
            returns.append(ret)
            posRisky = 0
        else:
            safeSell = safePrice
            ret = (safeSell - safeBuy)/safeBuy
            print("sell safe price: " + str(safeSell))
            print("safe position return: " + str(ret))
            print("="*15)
            returns.append(ret)
            posSafe = 0


print("return list: {}".format(returns))
```
